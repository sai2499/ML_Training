```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_lfw_people
import numpy as np
from sklearn.cluster import DBSCAN
```

```python
people = fetch_lfw_people(min_faces_per_person=20, resize=0.7)
image_shape = people.images[0].shape
fig, axes = plt.subplots(2,5,figsize=(15,8),subplot_kw={'xticks':(),'yticks': ()})
for target, image, ax in zip(people.target, people.images, axes.ravel()):
    ax.imshow(image)
    ax.set_title(people.target_names[target])
```

```python
print("People, images, shape: ", format(people.images.shape))
print("Number of classes: ", format(len(people.target_names)))
```

```python
#count how ofter each target appears
counts = np.bincount(people.target)
#print name next to each target name
for i, (count, name) in enumerate(zip(counts, people.target_names)):
    print("{0:25} {1:3}".format(name, count), end= ' ')
    if(i+1)%3 ==0:
        print()
    
```

```python
# to make data less skewed, will take up to 50 images of each person
mask = np.zeros(people.target.shape, dtype= bool)
for target in np.unique(people.target):
    mask[np.where(people.target == target)[0][:50]] =1
X_people = people.data[mask]
y_people = people.target[mask]

# scale the grayscale value to be between 0 and 1
# instead of 0 and 255 for better numeric stability
X_people = X_people /255
```

```python
pca = PCA(n_components=100, whiten=True, random_state=0)
pca.fit_transform(X_people)
X_pca=pca.transform(X_people)
```

```python
#======================= USing DBSCAN =========================================
dbscan= DBSCAN()
labels = dbscan.fit_predict(X_pca)
print("Unique labels: {} ".format(np.unique(labels)))
```

```python
"""
As labels are -1 entire data is labeled as "noise" by DBSCAN.
We can use two thing to help here: 
    a. Use higher eps to expand the neighborhood of each point
    b. set min_samples lower to consider smaller groups of points as clusters
"""
```

```python
#===========Changing min_samples first============
dbscan_1= DBSCAN(min_samples=3)
labels_1 = dbscan_1.fit_predict(X_pca)
print("Unique labels: {} ".format(np.unique(labels_1)))
# even changing min_samples everything is labeled as noise
```

```python
#==================Increasing eps===================
dbscan_2= DBSCAN(min_samples=3,eps=15)
labels_2 = dbscan_2.fit_predict(X_pca)
print("Unique labels: {} ".format(np.unique(labels_2)))
#using eps 15 we get single cluster and rest noise
```

```python
#count number of points in all clusters and noise.
#bin count doesnt allow negative numbers , so we need to add 1
#The first number in the result corresponds to noise points
print("Number of points per cluster: {}".format(np.bincount(labels+1)))
```

```python
noise = X_people[labels==-1]
fig, axes=plt.subplots(3,9, subplot_kw={'xticks': (), 'yticks': ()}, figsize=(12,4))
for image, ax in zip(noise, axes.ravel()):
    ax.imshow(image.reshape(image_shape), vmin=0, vmax=1)
```

```python
'''
1. The images contain odd angles or crops that are too close or wide
2. The method to find "the odd one out" is called outlier detection.
3. There is little we can do about people in photos sometimes wearfgn hats, drinking and so on, but it's good to know that
these are issues in the data that any algorithm we might apply needs to handle
'''
```

```python
for eps in [1,3,5,7,9,11,13]:
    print("eps ={}".format(eps))
    dbscan_2= DBSCAN(min_samples=3,eps=eps)
    labels_2 = dbscan_2.fit_predict(X_pca)
    print("Unique labels: {} ".format(np.unique(labels_2)))
    print("Cluster sizes: {}".format(np.bincount(labels_2+1)))
```

```python
'''
Lower setting of eps all are noise
eps=7 many noise and many small clusters
eps=9 many noise , one big cluster and some smaller clusters
eps=11 one large cluster and noise

Things to notice:
1. Atmost one large cluster containing most of the points and some smaller clusters
2. This indicates that there are not 2 or 3 different kinds of face images in tghe data that are very distinct, but rather
all images are more or less equally similar to the rest
'''
```

```python
dbscan_5 = DBSCAN(min_samples=3, eps=7)
labels = dbscan_5.fit_predict(X_pca)
for cluster in range(max(labels)+1):
    mask = labels ==cluster
    n_images = np.sum(mask)
    fig, axes = plt.subplots(1, n_images, figsize=(n_images*1.5,4), subplot_kw={'xticks':(),'yticks': ()})
    for image, label, ax in zip(X_people[mask], y_people[mask], axes):
        ax.imshow(image.reshape(image_shape), vmin=0, vmax=1)
        ax.set_title(people.target_names[label].split()[-1])
```
