```python
import mglearn
mglearn.plots.plot_agglomerative_algorithm()
```

```python
'''
1. Initially each point is its own cluster. Then, in each step, the two clusters that are closest are merged.
2. In the first four steps, two single point clusters are picked and these are koined into two point clusters
3. In Step 5, one of the two point clusters is extended to a thied point, and so on.
4. In Step 9, there are only three cluster remaining. After three clusters the algorithm stops.
5. Because of the way the algorithm works, agglomerative clustering cannot make prediction for new data points.So it has no 
predict method. To build a model and get the clustering membership on the training set, use the fit_predict method.
'''
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
X, y= make_blobs(random_state=1)

agg=AgglomerativeClustering(n_clusters=3)
assignment = agg.fit_predict(X)

mglearn.discrete_scatter(X[:,0],X[:,1],assignment)
plt.legend(["Cluster 0", "Cluster 1", "Cluster 2"], loc="best")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```
