```python
from scipy.cluster.hierarchy import dendrogram, ward
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt 
```

```python
X, y= make_blobs(random_state=0, n_samples=12)
#Apply ward clustering to the data array x
#The scipy ward function returns an array that specifies the distances
#bridged when performing agglomerative clustering
linkage_array = ward(X)
#now plot the dendrogram fro the linkage_array containing the distance between clusters
dendrogram(linkage_array)
ax=plt.gca()
bounds=ax.get_xbound()
ax.plot(bounds, [7.25, 7.25],'--',c='k')
ax.plot(bounds, [4,4],'--',c='k')
ax.text(bounds[1], 7.25, 'two clusters', va="center", fontdict={'size':15})
ax.text(bounds[1], 4, 'three clusters', va="center", fontdict={'size':15})
plt.xlabel("Sample index")
plt.ylabel("Cluster distance")
```

```python
'''
1. The dendrogram shows data points as points on the bottom.
2. Then, a tree is plotted with these points as the leaves, and a new node parent in added for each two cluster that are joined
3. The y-axis in the dendrogram doesn't specify when is the agglomerative algorithm two clusters get merged.
4. The lenght of each branch also shows how far apart the merge cluster are. 
5. The longest branch on the dendrogram are the three lines that are marked by the dashed line as three clusters
6. Unfortunately, agglomerative clustering still fails at separating complex shapes like the two_moons data.
'''
```
