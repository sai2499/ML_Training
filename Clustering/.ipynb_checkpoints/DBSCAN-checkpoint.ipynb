```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
X, y= make_blobs(random_state=0, n_samples=12)
dbscan= DBSCAN()
clusters = dbscan.fit_predict(X)
print("Cluster membership ", format(clusters))
```

```python
'''
All datapoints were assigned the label -1 which is noise. This is a consequence of the default parameters
setting for eps and min_samples, which are not tuned for small toy datasets. 
'''
```

```python
import mglearn
mglearn.plots.plot_dbscan()
```

```python
'''
1. In the plot, points that belong to clusters are solid, while the noise points are shown in white.
2. Core samples are shown as large markers, while boundary points are displayed as smaller markers.
3. Increasing eps, means that more points will be included in a cluster. This makes cluster grow, but 
might also lead to multiple cluster joining into one.
4. Increasing min_samples mean that fewer points will be core points and more points will be noise.
'''
```

```python
'''
1. eps is somewhat more important, as it determines what it means for points to be close.
2. Setting eps to be very small will mean that no points are core samples, and may lead to all points
as noise.
3. Whereas setting eps to very large will result in all points forming a single cluster
'''
```

```python
'''
1. Min_samples setting mostly determines whether points in less dense region will be labeled as outliers
or as their own cluster.
2. If we increase min_samples, anything that would have been a cluster with less than min_samples
many samples will now be noise.
3. min_sample detemines the minimum cluster size. As in the plot above: when going from min_sample=3 
to min_samples=5 with eps=1.5. 
    a. When min_samples=3 there are three clusters: one of four points, one of five points and one of
    three points.
    b. When min_samples=5, the two smaller clusters are now labelled noise, and only cluster with 
        5 samples remain.
'''
```

```python
'''
DBSCAN doesnt require setting the number of cluster explicitly, setting eps implicitly controls how 
many clusters will be found. Finding a good setting for eps is easier by scaling the data using 
standardScaler or minmaxscaler, this will ensure that all features have similar ranges.
'''
```

```python
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
X_1, y_1 = make_moons(n_samples =200, noise=0.05, random_state=0)
scaler = StandardScaler()
scaler.fit(X_1)
X_scaler = scaler.transform(X_1)
dbscan = DBSCAN(eps=0.5)
clusters_1 = dbscan.fit_predict(X_scaler)
plt.scatter(X_scaler[:,0], X_scaler[:,1],c=clusters_1, cmap=mglearn.cm2, s=60)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```
