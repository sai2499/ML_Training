```python
'''Using PCA and NMF there is nothing much can do to this data, as it lives in only two dimensions.
Reducing the data to 1D with PCA and NMF would destroy the data.
However, we can find more expressive representation with kmeans by using cluster center'''
```

```python
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt 
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
kmeans = KMeans(n_clusters=10, random_state=0)
kmeans.fit(X)
y_pred=kmeans.predict(X)
plt.scatter(X[:,0],X[:,1], c=y_pred,s=60, cmap="Paired")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=60, marker='^',c=range(kmeans.n_clusters),linewidth=2, cmap="Paired")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
print("Cluster Membership",format(y_pred))
```

```python
'''
1. We used 10 cluster centers, which means each point is now assigned a new number b/t 0 and 9.
2. It is also visible that data being represented using 10 components with all being 0, apart from the one that represents the
cluster center the point is assigned to.
3. Using this 10D representation, it would be possible to separate the 2 half moons using linear model, which would not been possible
using just two feature(upper and lower moon)
'''
```

```python
distance_features=kmeans.transform(X)
print("Distance feature shape: ", format(distance_features.shape))
print("Distance feature: ",format(distance_features))
'''
We get more expressive representation of the data by using the distance to each of the cluster center as features.
'''
```
