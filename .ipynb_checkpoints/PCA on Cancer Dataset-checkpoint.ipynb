```python
#as the cancer dataset has 30 features which would result in 430 Scatterplots.So we will draw histogram of each feature
#for the two cases(benign & malignant cancer)
import matplotlib.pyplot as plt
import mglearn
from sklearn.datasets import load_breast_cancer
import numpy as np
cancer = load_breast_cancer()
fig, axes = plt.subplots(15,2, figsize=(10,20))
malignant = cancer.data[cancer.target == 0]
benign = cancer.data[cancer.target == 1]

ax = axes.ravel()

for i in range(30):
    _, bins = np.histogram(cancer.data[:, i], bins=50)
    ax[i].hist(malignant[:,i], bins=bins, color= mglearn.cm3(0), alpha=.5)
    ax[i].hist(benign[:,i], bins=bins, color=mglearn.cm3(2), alpha=.5)
    ax[i].set_title(cancer.feature_names[i])
    ax[i].set_yticks(())
ax[0].set_xlabel("Feature Magnitude")
ax[0].set_ylabel("Frequency")
ax[0].legend(["malignant","benign"], loc="best")
fig.tight_layout()
```

```python
"""
This contains the histgram for each feature, counting how often a data point appears with a feature in a certain range
(called bins). This gives us some idea of how each feature is distributed across the two classes and allows us to 
venture a guess as to which feature is better distinguished.
THis plot doesnt show us anything about the interactions between variables and how these relate to the classes.
Using PCAm we can capture the main interactions and get a better picture. Now find the two PCA and visualize the data
in a 2d scatter plot.
"""
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(cancer.data)
X_scaled= scaler.transform(cancer.data)
```

```python
"""
Instantiate the PCA object find the principal components by calling fit method, then apply the rotation and dimensional
reduction by calling transform. PCA only rotates the data, but keeps all principal components.
To reduction the dimensionality of the data, we need to specify how many components we want to keep to create PCA object
"""
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components =2)
pca.fit(X_scaled)
X_pca = pca.transform(X_scaled)
print("Original Shape", format(X_scaled.shape))
print("Reduced Shape", format(str(X_pca.shape)))
```

```python
plt.figure(figsize=(8,8))
mglearn.discrete_scatter(X_pca[:,0], X_pca[:,1], cancer.target)
plt.legend(cancer.target_names, loc="best")
plt.gca().set_aspect("equal")
plt.xlabel("First PCA")
plt.ylabel("Second PCA")
```

```python
"""
PCA being an unsupervised method, it does not use any data information when finding the rotation. It simply looks at 
correlations in the data.
From the scatter plot it seems both the classes are seperated quite wellin 2d. This leads us to believe that even a 
linear classifier could do a reasonable job in distinguishing the two classes.
A downside of PCA is that the two axes in the plot are often not very easy to interpret.The PCA correspond to directions
in the original data, so they are combinations of original features. The principal components are generally stored in
the components_attribute of the PCA object during fitting
"""
```

```python
print("PCA component shape: ", format(pca.components_.shape))
```

```python
"""each row in components_ corresponds to on principal component, and they are sorted by their importance."""
```

```python
#Contents of components_ 
print("PCA components: ", format(pca.components_))
```

```python
plt.matshow(pca.components_, cmap="viridis")
plt.yticks([0,1], ["First Component", "Second Component"])
plt.colorbar()
plt.xticks(range(len(cancer.feature_names)), cancer.feature_names, rotation=60, ha = "left")
plt.xlabel("Feature")
plt.ylabel("Principal components")
```

```python
"""
All features have the same sign. That means that there is a general correlation between all feature. As on measurement
is high, the others are likely to be high as well. The second component has mixed signs, and both of the components
involve all of the 30 feature. This mixing of all feaures makes explaining the heat map tricky
"""
```
