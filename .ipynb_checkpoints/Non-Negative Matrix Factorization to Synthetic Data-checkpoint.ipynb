```python
'''
1. On the left, for NMF with 2 components it is clear that all points in the data can be written as positive combination of the 
two components. If there are enough components to perfectly reconstruct the data, the algorithm will choose directions that
point towards the extremes of the data.

2. On the right, NMF creates a component that points towards the mean, as pointing there best explains the data.
In contrast with PCA, reducing the number of components not only removes some direction, but creats an entirely different
set of components

3. Components in NMP are also not ordered in any specific way so all components play an equal part.

4. NMF uses a random initialization, which might lead to different results depending on the random seed.

5. In relatively simple cases such as a the synthetic data with two component, where all the data can be explained prefectly, the
randomness has little effect.
'''
```
